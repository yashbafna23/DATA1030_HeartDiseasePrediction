```python
pd.value_counts(df['cp']).plot.bar()
plt.ylabel('Count')
plt.xlabel('Chest Pain Types')
plt.show()
```

```python
df.plot.scatter('age','chol',alpha=1,s=50) 
plt.show()
```

```python
count_matrix = df.groupby(['sex', 'fbs']).size().unstack()
#print(count_matrix)

count_matrix_norm = count_matrix.div(count_matrix.sum(axis=1),axis=0)
print(count_matrix_norm)
```

```python
count_matrix_norm.plot(kind='bar', stacked=True)
plt.ylabel('FBS')
plt.legend(loc=4)
plt.show()
```

```python
import matplotlib
from matplotlib import pylab as plt

categories = df['fbs'].unique()
bin_range = (df['age'].min(),df['age'].max())

for c in categories:
    plt.hist(df[df['fbs']==c]['age'],alpha=0.5,label=c,range=bin_range,bins=20,density=True)
plt.legend()
plt.ylabel('fbs')
plt.xlabel('age')
plt.show()
```

```python
pd.plotting.scatter_matrix(df.select_dtypes(int), figsize=(9, 9), marker='o',hist_kwds={'bins': 50}, 
                           s=30, alpha=.1)
plt.show()
```
